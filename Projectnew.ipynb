```python
df.drop(['customer_id'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
out=ohe.fit_transform(df[['country']])
out
```

```python
df1=pd.DataFrame(out,columns=ohe.get_feature_names_out())
df1
```

```python
data=pd.concat([df,df1],axis=1)
data
```

```python
data.drop(['country'],axis=1,inplace=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
labelfit=le.fit(data['gender'])
data['gender']=labelfit.transform(data['gender'])
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for i in data.columns:
  sns.boxplot(x=data[i])
  plt.show()
```

```python
x=data.drop(['churn'],axis=1)
y=data['churn']
```

```python
#standaredscaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
y_train.value_counts()
```

```python
#smote
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_train_smote,y_train_smote=smote.fit_resample(x_train,y_train)
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train_smote,y_train_smote)
y_pred=knn.predict(x_test)
```

```python
y_pred
```

```python
training_score=knn.score(x_train_smote,y_train_smote)
testing_score=knn.score(x_test,y_test)
print(training_score)
print(testing_score)
```

```python
#evaluate
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#logistic_regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train_smote,y_train_smote)
y_pred=lr.predict(x_test)
```
